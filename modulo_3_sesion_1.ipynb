#### Carga y exploración inicial de los datos.
Inicialmente exploraremos la información disponible de años de estudio, experiencía, tener diplomado y nivel salarial.  

````python
## importación de librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
df = pd.read_csv('salarios_regresion.csv')
df.head()
````

#### Información general del dataset
A continuación se presenta la estructura de los datos.

````python
df.describe()
````

#### Visualización de relaciones entre las variables independientes y el salario.

A continuación se observa el comoportamiento de cada uno de las variables. Para ello se presenta el comportamiento de cada una de las variables respecto a la variable del salario.

````python
## gráficos dispersión de las variables de salario
sns.pairplot(df, x_vars=['Educacion', 'Experiencia', 'Tiene_diplomado'], y_vars='Salario', height=4, aspect=1, kind='reg')
plt.show()
````

#### Modelo de regresión lineal simple (con años de educación)

A continuación, se presenta el caso del modelo de regresión para evaluar el salario en función de los años de educación.

````python
## librerías requeridas para el modelo de regresión lineal
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
````

Se selecciona la variable independiente **X** (educación) y la variable dependiente del salario **Y**

````python
## selección de las variables
X_simple = df[['Educacion']]
y = df['Salario']
````

````python
## observación de los datos de educación - almacenados en variable "X_simple"
X_simple.hist()
````

````python
## comportamiento de la variable de salario almacenada en variable "y"
y.hist()
````

````python
## creación del modelo y entrenamiento
modelo_simple = LinearRegression()
modelo_simple.fit(X_simple, y)
modelo_simple
````

````python
# Predicciones, usando el modelo entrenado
y_pred_simple = modelo_simple.predict(X_simple)
## comportamiento de la variable de respuesta
plt.hist(y_pred_simple)
````

````python
## conversión de la predicción a un dataframe para observar sus medidas descriptivas
y_pred_simple = pd.Series(y_pred_simple)
y_pred_simple.describe()
````

A continuación se observa la gráfica de la predicción vs los valores reales.

````python
# Gráfico
plt.scatter(X_simple, y, label='Datos reales')
plt.plot(X_simple, y_pred_simple, color='red', label='Modelo lineal')
plt.xlabel('Años de educación')
plt.ylabel('Salario')
plt.title('Regresión lineal simple')
plt.legend()
plt.show()

## Valores del modelo de regresión
print(f"Coeficiente: {modelo_simple.coef_[0]:.2f}")
print(f"Intercepto: {modelo_simple.intercept_:.2f}")
````

Para comprender un poco más como esta funcionando el modelo propuesto, en este caso tendríamos la ecuación

$$
Salario = 113.40 Años_.de_.educación + 2318.88
$$

Ahora se procederá a construir un modelo completo, el que incluya todas las variables, para conocer su impacto sobre el salario.
