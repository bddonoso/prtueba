#### Modelo de regresin lineal simple (con aos de educacin)

A continuacin, se presenta el caso del modelo de regresin para evaluar el salario en funcin de los aos de educacin.

````python
## libreras requeridas para el modelo de regresin lineal
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
````

Se selecciona la variable independiente **X** (educacin) y la variable dependiente del salario **Y**

````python
## seleccin de las variables
X_simple = df[['Educacion']]
y = df['Salario']
````

````python
## observacin de los datos de educacin - almacenados en variable "X_simple"
X_simple.hist()
````

````python
## comportamiento de la variable de salario almacenada en variable "y"
y.hist()
````

````python
## creacin del modelo y entrenamiento
modelo_simple = LinearRegression()
modelo_simple.fit(X_simple, y)
modelo_simple
````

````python
# Predicciones, usando el modelo entrenado
y_pred_simple = modelo_simple.predict(X_simple)
## comportamiento de la variable de respuesta
plt.hist(y_pred_simple)
````

````python
## conversin de la prediccin a un dataframe para observar sus medidas descriptivas
y_pred_simple = pd.Series(y_pred_simple)
y_pred_simple.describe()
````

A continuacin se observa la grfica de la prediccin vs los valores reales.

````python
# Grfico
plt.scatter(X_simple, y, label='Datos reales')
plt.plot(X_simple, y_pred_simple, color='red', label='Modelo lineal')
plt.xlabel('Aos de educacin')
plt.ylabel('Salario')
plt.title('Regresin lineal simple')
plt.legend()
plt.show()

## Valores del modelo de regresin
print(f"Coeficiente: {modelo_simple.coef_[0]:.2f}")
print(f"Intercepto: {modelo_simple.intercept_:.2f}")
````

Para comprender un poco ms como esta funcionando el modelo propuesto, en este caso tendramos la ecuacin
