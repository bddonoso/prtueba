### **3. Manejo de fechas**

```python
# libreria para manipulación de fechas
import datetime as dt
```

Creación de nuevas columnas en el dataframe

```python
# creación de columna "Hora" donde almacena la hora de la fecha
Data1["Hora"] = pd.to_datetime(Data1["LastSeen"]).dt.hour
```

```python
# creación de columna "minutos" donde almacena los minutos de la fecha
Data1["Minuto"] = pd.to_datetime(Data1["LastSeen"]).dt.minute
```

```python
Data1.head(3)
```

```python
 ## creación de un nueva columna con el cálculo de la hora exacta
Data1["HoraExacta"] = Data1["Hora"] + Data1["Minuto"]*(1/60)
Data1.head(3)
```

Creación de un gráfico para el análisis de la información

```python
import matplotlib.pyplot as plt
## ggplot es un estilo para la presentación de la gráfica
plt.style.use("ggplot")

## datos y configuración de la gráfica (scatter plot)
ax = Data1.plot(x="LastSeen", y="HoraExacta", style='.', color='b', legend=False)

## titulo
plt.title("Hora de Desconexión VS. Fecha")

## manipulación eje x - rotation set to 45
ax.tick_params(axis='x', rotation=45)

## adición del nombre del eje y
ax.set_ylabel("Hora Exacta")

## comando para mostrar la gráfica
plt.show()
```

### **4. Extracción de información de columnas**

Se analizará la información del uso del celular, para identificar diferentes patrones de uso

________________________________________________________________

Aplicación de funciones **.apply**

```python
## dirección de información
file = "Fac_Telefono.csv"

# lectura de archivo
Cel = pd.read_csv(file)
Cel.head(3)
```

```python
# funciones sobre columnas
# se crea una nueva columna extrayendo los carácteres del 10 al 11 de la duración,
# que corresponden a los minutos que dura la llamada

Cel["DuracionOK"] = Cel["Duracion"].apply(lambda x: x[10:12])
```

_______________________________________________________________
### **5. Formato de columnas**
Conversión de las columnas al formato adecuado

```python
Cel.info()
```

```python
# Convertir a string
# (para comparar con Departamentos y Municipios del celular)

Cel["CodeAntena"] = Cel["CodeAntena"].astype(str)
Cel["Hora"] = pd.to_datetime(Cel["Hora"])
Cel["DuracionOK"] = Cel["DuracionOK"].astype(int)
```

```python
Cel.info()
```

```python
#agrega columnas de dia y mes
#columna dia .weekday es un método que arroja el número del día 4 es
Cel["dia"] = Cel["Hora"].apply(lambda x: x.weekday())

#.month calcula el numero del mes, pero como atributo
Cel["mes"] = Cel["Hora"].apply(lambda x: x.month)
Cel.head(3)
```

Revisar categorias existentes en una fila

```python
## value_counts muy funcional para análizar variables categóricas
Cel["mes"].value_counts()
```

```python
# Usar dias y meses en español como referencia
# renombrar valores en las columnas
#creación de diccionarios, para asociar numeros a días y meses
dias = {0:"1. Lun",1:"2. Mar",2:"3. Mie",3:"4. Jue",4:"5. Vie",5:"6. Sab",6:"7. Dom"}
## solo se utilizan estos dado que solo tenemos meses de 6 al 8
meses = {6:"Jun",7:"Jul",8:"Ago",9:"Sep"}

Cel["dia"] = Cel["dia"].apply(lambda x: dias[x])
Cel["mes"] = Cel["mes"].apply(lambda x: meses[x])

Cel.head()
```

### **6. Remover columnas**

Se eliminan todas las columnas que no son necesarias

```python
# eliminación columnas
Cel.drop( ["Unnamed: 0", "CodeAntena", "Duracion"], axis=1, inplace=True)
Cel.head(3)
```

## **7. Sintesis de datos**

Utilización del comando Groupby

```python
# GROUPBY: Muy importante para análisis.
## Presenta el promedio de los tiempos de llamada por día de la semana
Cel[["dia","DuracionOK"]].groupby(["dia"]).mean()
```

```python
## Presenta los tiempos totales de llamadas por mes de la semana
Cel[["mes","DuracionOK"]].groupby(["mes"]).sum()
```

```python
#agrupaciones por mes y dia de la semana
Aux = Cel[["dia","mes","DuracionOK"]].groupby(["mes","dia"]).mean()
Aux
```

```python
# Stack y Unstack: importante al trabajar con multi-indices
# (i.e., dimensiones altas)
    # Stack es enviar columnas a indices
    # Unstack es enviar indices a columnas

# Unstack es enviar indices a columnas
Aux = Aux.unstack(level=1)
Aux

#Aux = Aux.stack()
```

```python
#quita el nombre (valor) porque ahora todo es de valor
Aux.columns = Aux.columns.droplevel()
Aux
```

Veámoslo graficamente

```python
#método listo para gráficar
Aux.plot(kind='bar')
plt.title("tiempos de llamada promedio agrupado por 'dia de la semana' Y 'mes'")
plt.show()
```
