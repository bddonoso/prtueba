```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

#Crear un dataset simulado
data = {
    'IngresoMensual':[3000, 1500, 4000, 1200, 2500, 800, 5000, 2000],
    'Edad':[25, 40, 35, 28, 50, 22, 45, 30],
    'HistorialPago':[1, 0, 1, 0, 1, 0, 1, 0], #1 = buen historial, 0 = mal historial
    'CreditoAprobado':[1, 0, 1, 0, 1, 0, 1, 0]
}

df = pd.DataFrame(data)
df
```

```python
#Separamos variables predictoras(X) y la variable objetivo(y)
X = df[['IngresoMensual', 'Edad', 'HistorialPago']]
y = df['CreditoAprobado']

#Crear el modelo
modelo = DecisionTreeClassifier(max_depth=3, random_state=42)

#Entrenar el modelo
modelo.fit(X, y)
```

```python
#Visualizar arbol
plt.figure(figsize=(10, 8))
plot_tree(modelo, feature_names=X.columns, class_names=['No Aprobado', 'Aprobado'], filled=True, rounded=True)
plt.show()
```

```python
#Hacer predicciones
nuevos_clientes = pd.DataFrame({
    'IngresoMensual':[1800, 4500],
    'Edad':[30, 45],
    'HistorialPago':[0, 1]
})

predicciones = modelo.predict(nuevos_clientes)
nuevos_clientes['CreditoAprobado'] = predicciones
print(predicciones)
```

```python
#Evaluamos el modelo
accuracy = modelo.score(X, y)
print(f'Precisión del modelo: {accuracy * 100:.2f}%')
```
